```python
import numpy as np
import pandas_datareader.data as reader
import pandas as pd
import datetime as dt
import statsmodels.api as sm
```

```python
end = dt.date(2020,12,31)
start = dt.date(end.year - 5, end.month, end.day)
```

```python
funds= ['PRGO', 'CNC', 'CVS', 'CI', 'UHS', 'TAP', 'SJM', 'ADM', 'TSN', 'COTY', 'PVH', 'CCL', 'MHK', 'F', 'KSS', 'XRX', 'NLOK', 'FIS', 'DXC', 'WDC', 'MRO', 'HFC', 'NOV', 'FANG', 'KMI']
```

```python
fundsret = reader.get_data_yahoo(funds,start,end)['Adj Close'].pct_change()
```

```python
fundsret_mtl = fundsret.resample('M').agg(lambda x: (x+1).prod()-1)
```

```python
fundsret_mtl = fundsret_mtl[1:]
```

```python
fundsret_mtl
```

```python
factors = reader.DataReader('F-F_Research_Data_Factors','famafrench',start,end)[0]
```

```python
factors = factors[1:]
```

```python
factors
```

```python
fundsret_mtl.shape
```

```python
fundsret_mtl.index = factors.index
```

```python
fundsret_mtl
```

```python
merge = pd.merge(fundsret_mtl,factors,on='Date')
```

```python
merge
```

```python
merge[['Mkt-RF','SMB','HML','RF']] = merge[['Mkt-RF','SMB','HML','RF']]/100
```

```python
merge
```

```python
merge['PRGO-RF'] = merge.PRGO - merge.RF
merge['CNC-RF'] = merge.CNC - merge.RF
merge['CVS-RF'] = merge.CVS - merge.RF
merge['CI-RF'] = merge.CI - merge.RF
merge['UHS-RF'] = merge.UHS - merge.RF
merge['TAP-RF'] = merge.TAP - merge.RF
merge['SJM-RF'] = merge.SJM - merge.RF
merge['ADM-RF'] = merge.ADM - merge.RF
merge['TSN-RF'] = merge.TSN - merge.RF
merge['COTY-RF'] = merge.COTY - merge.RF
merge['PVH-RF'] = merge.PVH - merge.RF
merge['CCL-RF'] = merge.CCL - merge.RF
merge['MHK-RF'] = merge.MHK - merge.RF
merge['F-RF'] = merge.F - merge.RF
merge['KSS-RF'] = merge.KSS - merge.RF
merge['XRX-RF'] = merge.XRX - merge.RF
merge['NLOK-RF'] = merge.NLOK - merge.RF
merge['FIS-RF'] = merge.FIS - merge.RF
merge['DXC-RF'] = merge.DXC - merge.RF
merge['WDC-RF'] = merge.WDC - merge.RF
merge['MRO-RF'] = merge.MRO - merge.RF
merge['HFC-RF'] = merge.HFC - merge.RF
merge['NOV-RF'] = merge.NOV - merge.RF
merge['FANG-RF'] = merge.FANG - merge.RF
merge['KMI-RF'] = merge.KMI - merge.RF
```

```python
merge
```

```python
y = merge['PRGO-RF']
X = merge[['Mkt-RF','SMB','HML']]
X_sm = sm.add_constant(X)
model = sm.OLS(y,X_sm)
results = model.fit()
results.summary()
```
